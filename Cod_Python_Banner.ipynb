```
import mysql.connector 
import matplotlib.pyplot as plt
lista1 = []
lista2 = []
lista3 = []
lista4 = []
lista5 = []
lista6 = []
cnx = mysql.connector.connect(
    #host and password 
    database = 'enem' )

variavel = cnx.cursor()
consultar = 'select count(NU_IDADE), ROUND((AVG(NU_NOTA_CN)+AVG(NU_NOTA_CH)+AVG(NU_NOTA_LC)+AVG(NU_NOTA_MT)+AVG(NU_NOTA_REDACAO))/5)from enem.microdados_enem_2017 where NU_IDADE>15 and NU_IDADE<=60 and TP_PRESENCA_CN=1 and TP_PRESENCA_CH=1 and TP_PRESENCA_LC=1 and TP_PRESENCA_MT=1 group by Nu_idade order by NU_IDADE'
variavel.execute(consultar)
dados = variavel.fetchall()

#print(dados)
for i in range (len(dados)):
  for j in range (3):
    if j==0:
      lista1.append(dados[i][j])
    if j==1:
      lista2.append(dados[i][j])

for i in range (16,61):
  lista3.append(i)

for i in lista3:
  i=i+0.45/2
  lista4.append(i)
    
for a in lista3:
    a=a-0.45/2
    lista6.append(a)

for i in lista1:
  i=((i*100)/185000)
  lista5.append(i)
  
from pylab import rcParams
rcParams['figure.figsize'] = 20, 7

fig, ax1 = plt.subplots()

ax1.bar(lista4,lista2,width=0.45,color='g')
plt.axis([15, 61, 490, 560])
plt.ylabel('Notas dos participantes')
plt.xlabel('Idade dos participantes')

ax2 = ax1.twinx()
ax2.bar(lista6,lista5,width=0.45,color='r')
ax2.axis([15, 61, 0, 20])
plt.ylabel('Porcentagem de participantes')

plt.legend(loc='upper right')
plt.show()
```
